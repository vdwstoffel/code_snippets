Pre Setup

```python
import pandas as pd                                                             # pip install pandas
df = pd.read_csv('QueryResults.csv', names=['DATE', "TAG", "POSTS"], header=0)  # Read from csv, giving custom column names and setting the heading as row 0
df = df.pivot(index='DATE', columns='TAG', values='POSTS')
df.fillna(0, inplace=True)
df.head()
```

Single line

```python
import pandas as pd                                    # pip install pandas
import matplotlib.pyplot as plt                        # pip install matplotlib

plt.plot(df.index, df['java'])
plt.xlabel("Date")                                      # x-axis label
plt.ylabel("Number of Posts")                           # y-axis label
plt.title("Java Posts")                                 # title of the graph
plt.show()
```

Multi line

```python
# The most straight forward way is just to call plot multiple times
plt.plot(df.index, df['java'], label='java')
plt.plot(df.index, df['python'], label='python')
plt.xlabel("Date")                                      # x-axis label
plt.ylabel("Number of Posts")                           # y-axis label
plt.title("Posts")                                      # title of the graph
plt.legend()                                            # add a legend to the graph
```

Rolling

```python
df = df.rolling(window=6).mean()

# The most straight forward way is just to call plot multiple times
plt.plot(df.index, df['java'], label='java')
plt.plot(df.index, df['python'], label='python')
plt.xlabel("Date")                                      # x-axis label
plt.ylabel("Number of Posts")                           # y-axis label
plt.title("Posts")                                      # title of the graph
plt.legend()                                            # add a legend to the graph
```
